```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
media_estados = pd.read_csv("df_estados.csv", sep=",", encoding='Utf-8', index_col=0) 
#index_col=0 avisamos que j√° existe uma coluna de index no arquivo
```

```python
media_estados.head(27)
```

```python
idhm=pd.read_csv('idhm.csv', sep=';', encoding='ISO_8859_1')
idhm.head(29)
```

```python
#Precisamos de uma coluna em comum nestes datasets para conseguirmos fazer o join entre eles
#Para isso, vou adicionar ao DF-IDHM uma coluna com as siglas dos estados
```

```python
dic_siglas={'Acre':'AC','Alagoas':'AL','Amazonas':'AM','Amap√°':'AP','Bahia':'BA','Cear√°':'CE',
           'Distrito Federal':'DF','Esp√≠rito Santo':'ES','Goi√°s':'GO','Maranh√£o':'MA','Minas Gerais':'MG',
           'Mato Grosso':'MT','Mato Grosso do Sul':'MS','Par√°':'PA','Para√≠ba':'PB','Pernambuco':'PE','Piau√≠':'PI',
           'Paran√°':'PR','Rio de Janeiro':'RJ','Rio Grande do Norte':'RN','Rond√¥nia':'RO','Roraima':'RR',
           'Rio Grande do Sul':'RS','Santa Catarina':'SC','Sergipe':'SE','S√£o Paulo':'SP','Tocantins':'TO','Brasil':'BR'}
```

```python
# Forma mais f√°cil de adicionar nova coluna √° um DF atrav√©s da fun√ß√£o map
idhm['SG_UF'] = idhm['Territorialidades'].map(dic_siglas)
idhm.head()
```

```python
enem_idhm=pd.merge(media_estados, idhm, how='left', left_on=['SG_UF_PROVA'], right_on=['SG_UF'])
# left_on=['SG_UF_PROVA'], right_on=['SG_UF'] √© a coluna que ambos datasets tem de informa√ß√£o em comum
```

```python
enem_idhm
```

```python
enem_idhm.rename(columns={'IDHM 2017': 'IDHM_2017', 'IDHM Educa√ß√£o 2017':'IDHM_Educ_2017',
                         'IDHM Longevidade 2017':'IDHM_Long_2017', 'IDHM Renda 2017':'IDHM_Renda_2017'}, inplace=True)
```

# CONVERTER UMA S√âRIE DO PANDAS (COLUNAS DF) DE OBJECT PARA FLOAT

```python
#CONVERTER UMA S√âRIE DO PANDAS (COLUNAS DF) DE OBJECT PARA FLOAT
# x: Pegue CADA ELEMENTO DA S√âRIE PANDAS em string e substitua as v√≠rgulas por pontos e transforme em float
#Estamos mudando , para . pois o df n√£o entendeu a , do n√∫mero
#lambda permite que a fun√ß√£o apply aplique oque pedimos (transformar cada x da s√©rie em float) para toda coluna


enem_idhm['IDHM_2017'] = enem_idhm['IDHM_2017'].apply(lambda x: float(str(x).replace(',', '.') ))
enem_idhm['IDHM_Educ_2017'] = enem_idhm['IDHM_Educ_2017'].apply(lambda x: float(str(x).replace(',', '.') ))
enem_idhm['IDHM_Long_2017'] = enem_idhm['IDHM_Long_2017'].apply(lambda x: float(str(x).replace(',', '.') ))
enem_idhm['IDHM_Renda_2017'] = enem_idhm['IDHM_Renda_2017'].apply(lambda x: float(str(x).replace(',', '.') ))
```

```python
enem_idhm.dtypes
```

```python
enem_idhm.plot.scatter('NU_NOTA_MT', 'IDHM_2017')
```

```python
enem_idhm.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017')
```

```python
media_redacao=enem_idhm.NU_NOTA_REDACAO.mean()
media_mat=enem_idhm.NU_NOTA_MT.mean()
media_idhm=enem_idhm.IDHM_2017.mean()
print('Media Reda√ß√£o =',media_redacao, 'M√©dia matem√°tica =',media_mat,'M√©dia IDHM =',media_idhm )
```

### GERANDO OS SCATTER PLOT COM MASCARAMENTO BOOLEANO

SELECIONAR COM [ ] A INFORMA√á√ÉO DO DF DE UMA EXPRESSAO BOOLEANA ( VERDADEIRO OU FALSO)

```python
q1=(enem_idhm.IDHM_2017 < media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
q2=(enem_idhm.IDHM_2017 > media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
q3=(enem_idhm.IDHM_2017 > media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)
q4=(enem_idhm.IDHM_2017 < media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)

fig, ax = plt.subplots(2,2, figsize=(10,6), )

media_enem_idhm_1=enem_idhm[q1]
media_enem_idhm_1.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [1,0]) 

media_enem_idhm_2=enem_idhm[q2]
media_enem_idhm_2.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [0,0])

media_enem_idhm_3=enem_idhm[q3]
media_enem_idhm_3.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [0,1])

media_enem_idhm_4=enem_idhm[q4]
media_enem_idhm_4.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [1,1])
```

### DEFINIR LIMITES DE X E Y DE CADA GR√ÅFICO

```python
q1=(enem_idhm.IDHM_2017 < media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
q2=(enem_idhm.IDHM_2017 > media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
q3=(enem_idhm.IDHM_2017 > media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)
q4=(enem_idhm.IDHM_2017 < media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)

fig, ax = plt.subplots(2,2, figsize=(10,6), )

min_redacao=enem_idhm.NU_NOTA_REDACAO.min()
max_redacao=enem_idhm.NU_NOTA_REDACAO.max()
min_idhm=enem_idhm.IDHM_2017.min()
max_idhm=enem_idhm.IDHM_2017.max()

media_enem_idhm_1=enem_idhm[q1]
media_enem_idhm_1.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [1,0], 
                               xlim = (min_redacao, media_redacao), ylim = (min_idhm, media_idhm)) #ESTABELECENDO LIMITES

media_enem_idhm_2=enem_idhm[q2]
media_enem_idhm_2.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [0,0], 
                               xlim = (min_redacao, media_redacao), ylim = (media_idhm, max_idhm))

media_enem_idhm_3=enem_idhm[q3]
media_enem_idhm_3.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [0,1], 
                               xlim = (media_redacao, max_redacao), ylim = (media_idhm, max_idhm))

media_enem_idhm_4=enem_idhm[q4]
media_enem_idhm_4.plot.scatter('NU_NOTA_REDACAO', 'IDHM_2017', ax = ax [1,1],
                               xlim = (media_redacao, max_redacao), ylim = (min_idhm, max_idhm))
```
